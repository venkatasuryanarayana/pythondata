```python
import pandas as pd

def convert(cell):
    if cell=='n.a.':
        return 'sam walton'
    return cell
def convert1(cell):
    if cell=="not available":
        return None
    return cell
    
df=pd.read_excel("write/stock_data.xlsx","Sheet1",converters = {
    'people':convert, # changed n.a. to sam walton
    'eps':convert1 # changed to not available to NaN
})
df
```

```python
df.to_excel("write/new.xlsx",sheet_name="stocks") # here above data stored in the file
```
